```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
df={}
for year in range(2000,2020):
    df[str(year)]=pd.read_csv('yearMetrics/tempMpUrlMetrics'+str(year)+'.csv')
    columns=['slno','word_count','body_text_percent','emphasized_body_text_percent','text_position_changes','text_cluster_count','link_count','page_size','graphic_percent','graphic_count','colour_count','font_count','colourfulness','visual_complexity','url']
    df[str(year)].columns=columns
```

```python
mean=np.zeros((20,13))
std=np.zeros((20,13))
for year in range(2000,2020):
    try:
        df[str(year)].drop(['slno','url'],1,inplace=True)
    except:
        pass
    try:
        yearMean=np.mean(df[str(year)].values,axis=0)
        yearStd=np.std(df[str(year)].values,axis=0)
    except:
        print(year)
    #print(yearMean)
    #print(res[year-2000])
    mean[year-2000]=yearMean
    std[year-2000]=yearStd
```
